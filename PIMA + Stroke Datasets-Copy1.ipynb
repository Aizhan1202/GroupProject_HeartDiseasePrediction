##### Tasks:

Rename Columns with matching names such as Age, BMI (can be left), Outcome & diabetes (hasDiabetes), glucose (Glucose), BloodPressure and diaBP (DiastolicBloodPressure).

```python
df_diabetes = df_pima.copy()
```

```python
df_diabetes.rename(columns={"Outcome": "HasDiabetes", "BloodPressure": "DiastolicBloodPressure"}, inplace=True)
```

```python
df_diabetes
```

Add gender column and set all values to female.

```python
df_diabetes['Gender'] = "Female"
```

Create column to identify dataset before merging.

```python
df_diabetes['Dataset'] = "PIMA"
```

```python
df_heart_disease
```

```python
new_column_names = {
    "male": "IsMale", "age": "Age", "education": "Education", "currentSmoker": "IsSmoking", "cigsPerDay": "NumCigsPerDay", 
    "prelaventStroke": "PrevalentStroke", "prevalentHyp": "PrevalentHyp","diabetes": "HasDiabetes",
    "totChol": "TotalCholesterol", "sysBP": "SystolicBloodPressure", "diaBP": "DiastolicBloodPressure",
    "heartRate": "HeartRate", "glucose": "Glucose"
}
df_2 = df_heart_disease.copy().rename(columns=new_column_names)
```

```python
df_2
```

Convert IsMale column to gender.

```python
def get_gender(row):
    if row['IsMale'] == 1:
        return "Male"
    else:
        return "Female"
df_2['Gender'] = df_2.apply(lambda row: get_gender(row), axis=1)
```

```python
df_2.drop(columns=['IsMale'], inplace=True)
```

```python
df_2['Dataset'] = "Heart Disease"
```

```python
df_2
```

Merge datasets, keep data that doesn't overlap for later analysis.

```python
df_diabetes = df_diabetes.merge(df_2, how="outer")
```

```python
df_diabetes
```

```python
def is_male(row):
    if row['Gender'] == "Male":
        return 1
    else:
        return 0
```

```python
df_diabetes['IsMale'] = df_diabetes.apply(lambda row: is_male(row), axis=1)
```

#### Modelling attempt 2

```python
x = df_training[df_training.columns[:-1]]
y = df_training['HasDiabetes']
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
#from xgboost import XGBClassifier
#from lightgbm import LGBMClassifier
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.metrics import accuracy_score, roc_auc_score, roc_curve, \
    classification_report
```

```python
models = [('LR', LogisticRegression()),
          ('KNN', KNeighborsClassifier()),
          ('CART', DecisionTreeClassifier()),
          ('RF', RandomForestClassifier()),
          ('SVR', SVC(gamma='auto')),
          ('GB',GradientBoostingClassifier())]

# evaluate each model in turn
results = []
names = []

for name, model in models:
    kfold = KFold(n_splits=10, random_state=None)
    cv_results = cross_val_score(model, x, y, cv=kfold, scoring="accuracy")
    results.append(cv_results)
    names.append(name)
    msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
    print(msg)
```

Top Models: LogisticRegression, Random Forest, Gradient Boosting.

```python
rf_model = RandomForestClassifier()
# Model Tuning
rf_params = {"n_estimators" :[100,200,500,1000], 
             "max_features": [3,5,7], 
             "min_samples_split": [2,5,10,30],
            "max_depth": [3,5,8,None]}
rf_cv_model = GridSearchCV(rf_model,
                           rf_params, 
                           cv=10, 
                           n_jobs=None, 
                           verbose=2).fit(x, y)
rf_cv_model.best_params_
# Final Model
rf_tuned = RandomForestClassifier(**rf_cv_model.best_params_).fit(x, y)
```

```python
# GBM
gbm_model = GradientBoostingClassifier()
# Model Tuning
gbm_params = {"learning_rate": [0.01, 0.1, 0.001],
               "max_depth": [3,5, 8, 10],
               "n_estimators": [200, 500, 1000],
               "subsample": [1, 0.5, 0.8]}
gbm_cv_model = GridSearchCV(gbm_model,
                            gbm_params,
                            cv=10,
                            n_jobs=None,
                            verbose=2).fit(x, y)
gbm_cv_model.best_params_
# Final Model
gbm_tuned = GradientBoostingClassifier(**gbm_cv_model.best_params_).fit(x,y)
```

```python
# LR
train, test = train_test_split(df_training, test_size=0.3, random_state=0, stratify=y)
train_x = train[train.columns[:-1]]
test_x = test[test.columns[:-1]]
train_y = train['HasDiabetes']
test_y = test['HasDiabetes']

lr_model = LogisticRegression()

lr_model.fit(train_x, train_y)

prediction = lr_model.predict(test_x)

metrics.accuracy_score(prediction, test_y)

metrics.confusion_matrix(prediction, test_y)
```

```python
models = [('RF', rf_tuned),
          ("LR", lr_model),
          ("GBM", gbm_tuned)]
results = []
names = []
for name, model in models:
    kfold = KFold(n_splits=10, random_state=None)
    cv_results = cross_val_score(model, x, y, cv=10, scoring="accuracy")
    results.append(cv_results)
    names.append(name)
    msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
    print(msg)
```
